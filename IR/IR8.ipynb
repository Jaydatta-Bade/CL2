#### 3. Write a program to construct a Bayesian network considering medical data. Use this model to demonstrate the diagnosis of heart patients using the standard Heart Disease Data Set (You can use Java/Python ML library classes/API.)

```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
# keeping only imp features
new_df = data[['cp', 'thalach', 'exang', 'oldpeak', 'ca', 'thal', 'target']]
new_df.head()
```

```python
# Define the structure of your Bayesian network
model = BayesianNetwork([
    ('cp', 'target'),
    ('thalach', 'target'),
    ('exang', 'target'),
    ('oldpeak', 'target'),
    ('ca', 'target'),
    ('thal', 'target')
])
```

```python
# Learn the parameters of the network from the data
model.fit(new_df, estimator=MaximumLikelihoodEstimator)
```

```python
# Perform inference
inference = VariableElimination(model)
```

```python
# You can now query the network for the probability of heart disease given some evidence
# Example evidence (you can replace this with actual patient data)
evidence = {
    'cp': data.at[0, 'cp'],
    'thalach': data.at[0, 'thalach'],
    'exang': data.at[0, 'exang'],
    'oldpeak': data.at[0, 'oldpeak'],
    'ca': data.at[0, 'ca'],
    'thal': data.at[0, 'thal']
}
```

```python
# Query the network
result = inference.query(variables=['target'], evidence=evidence)
print(result)
```
