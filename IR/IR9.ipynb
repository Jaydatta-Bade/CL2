#### 4. Implement Agglomerative hierarchical clustering algorithm using appropriate dataset.

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Generate sample data
X, labels_true = make_blobs(n_samples=90, centers=3, cluster_std=0.60, random_state=0)

# Plotting the generated data
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', s=50)
plt.grid()
plt.show()
```

```python
# Create a hierarchical clustering model
model = AgglomerativeClustering(affinity='euclidean', linkage='ward', n_clusters=3)

# Fit model to the data (no predictions needed, model contains clusters)
model.fit(X)
```

```python

# Plot the clusters
plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='rainbow', marker='o', label='Data points')
plt.title('Agglomerative Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 8))
Z = linkage(X, method='ward')
dendrogram(Z)
plt.title('Dendrogram for the clustering')
plt.xlabel('Data points')
plt.ylabel('Euclidean distances')
plt.show()
```
